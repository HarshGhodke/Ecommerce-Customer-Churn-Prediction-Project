### Separate x=Features (independent variable) & y=label (Dependent variable)

```python
x=df.drop(['churn'],axis=1)   # Selecting features by dropping churn column
y=df['churn']                 # Selecting label for traning as well as testing on a dataframe column for prediction
```

### Preprocessing

 Data 'preprocessing' is a process of preparing the raw data and making it suitable for a machine learning model. It is the first and crucial step while creating a machine learning model.

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
cat=[]                                     # In these step we seperate 'categorical' and 'continuous' columns in a dataframe by writing if__else statement
con=[]

for i in x.columns:
    if x[i].dtypes=='object':
        cat.append(i)
    else:
        con.append(i)
```

```python
cat
```

```python
con
```

### Pipe line

Pipeline--> Here we take pipeline approach which allows machine learning to scale

```python
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='median')),('scaler',StandardScaler())])
cat_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),('encode',OneHotEncoder())])
```

```python
pre=ColumnTransformer([('num_pipe',num_pipe,con),('cat_pipe',cat_pipe,cat)])   # pre=ColumnTransformer--> we used to combine two pipelines. In a column transformer we need to pass a list and also give the name of pipelines. 
```

```python
x1=pre.fit_transform(x)   # x1=pre.fit_transform(x)--> The above whole pipeline is fit to x
x1
```

```python
cols=pre.get_feature_names_out()  # cols=pre.get_feature_names_out() --> This step converts array into dataframe and also this step is used to take features.
cols
```

```python
x2=pd.DataFrame(x1,columns=cols)      # x2=pd.DataFrame(x1,columns=cols)--> In these step we made a dataframe & also pass a x1 array.so our preprocessed data is completed in these step.
x2
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split   
x_train,x_test,y_train,y_test=train_test_split(x2,y, test_size=0.2,random_state=21)   # Here we pass preprocessed x2 & target column y.
```

### Model Building

```python
from sklearn.linear_model import LogisticRegression    # We build multiple algorithms to find maximum f1_score from all the algorithms that we taken here.
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
```

```python
la=LogisticRegression()          # Here we create object  
kn=KNeighborsClassifier()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
ab=AdaBoostClassifier()
sv=SVC()
```

```python
list1=[la,kn,dt,rf,ab,sv]   # Here we take all the object in a list
```

```python
for i in list1:                         
    model=i.fit(x_train,y_train)
    
    y_pred=model.predict(x_test)
    
    from sklearn.metrics import f1_score,accuracy_score
    acc=accuracy_score(y_pred,y_test)
    f=f1_score(y_pred,y_test)
    
    print('*'*50)
    print(i)
    print('Accuracy:',acc)
    print('f1_score:',f)
    print('*'*50)
```

### Training Data Evaluation
